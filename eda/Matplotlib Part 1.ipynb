```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
#Lets create 2D diagrame
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='r')
plt.title('Graph in 2D')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.savefig('Test.png')
plt.show()
```

```python
y=x *x
```

```python
plt.plot(x,y,'r*-',linestyle='dashed',linewidth=4,markersize=12)
plt.title('Graph in 2D')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r*-')

plt.subplot(2,2,2)
plt.plot(x,y,'go-')

plt.subplot(2,2,4)
plt.plot(x,y,'r--')
```

```python
#compute x and y coordinates for points on a sine curve
x=np.arange(0,4 *np.pi,0.1)
y=np.sin(x)
plt.title('Sine wave form')
plt.plot(x,y)
plt.show()
```

```python
#subplot
#compute x and y coordinates for points on sine and cos curve
x=np.arange(0, 5*np.pi ,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#setup a subplot grid that has height 2, width 1 , 1 st active subplot
plt.subplot(2,1,2)

#Make the fist subplot
plt.plot(x,y_sin,'r--')
plt.show()

#set seconda subplot as active and make second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.show()
```

```python
#Bar Plot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.show()
```

```python
#Histograms - it gives count/colume/density of values on x axis
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

```python
#Boxplot-helps to find our percentiles

data=[np.random.normal(0,std,100) for std in range (1,4)]

#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
#Pie Chart

labels='Python','C++','Ruby','Java'
sizes=[235,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.4,0,0.1,0) #how much slide should be explot from other slices-explode 1st slice

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
