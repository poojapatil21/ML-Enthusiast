```python
import seaborn as sns
```

Distribution plots
------------

distplot

```python
df=sns.load_dataset('tips')
df.head()
```

Corelation with Heatmap
------
It uses colored cells , typically in a monochromatic scale, to show a 2D corelation matrix between two discrete dimetions or event types.It is important for feature selection.


```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

Join Plot
------
It allows to study the relationship between 2 numeric variables.The central chart display their correlation.

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

Pair Plot
---------
It is also known as scatterplot, in which one variable in the same data row is matched with another variable's value , like this: Pairs plots are just elaborations on this , showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

Dist Plot
-------
Helps us to check the distribution of the column features

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

Categorical Plots
---------
Boxplot   
Violinplot   
Countplot   
bar plot  


```python
#countplot - we ahve to give only one feature and other axis will show count 

sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot(y='day',data=df)
```

```python
#Bar Plot- we have to give two features for both axis.
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='total_bill',y='smoker',data=df)
```

```python
sns.barplot(y='total_bill',x='smoker',data=df)
```

```python
#Box Plot -Box and whisker plot is a graph that presents information in a percentile format.0,25,50,75,100
```

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')   #defualt it takes numerical features
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
#Violin Plot-it helps to see both distribution of data in terms of Kernel density estimation and the box plot
```

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
