```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('iris_dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.iris.unique()
```

Univariate Analysis
---------

```python
df_setosa=df.loc[df['iris']=='Iris-setosa']
df_versicolor=df.loc[df['iris']=='Iris-versicolor']
df_virginica=df.loc[df['iris']=='Iris-virginica']
```

```python
#Plot graphs of each variable for above dataframes as below:

plt.plot(df_setosa['sepal length'],np.zeros_like(df_setosa['sepal length']),'o',c='r')
plt.plot(df_versicolor['sepal length'],np.zeros_like(df_versicolor['sepal length']),'o',c='g')
plt.plot(df_virginica['sepal length'],np.zeros_like(df_virginica['sepal length']),'o',c='b')
plt.xlabel('Sepal Length Analysis')
plt.show()
```

BiVariate Analysis
------

```python
sns.FacetGrid(df,hue='iris',size=5).map(plt.scatter,'petal length','sepal width').add_legend()
plt.show()
```

MultiVariate Ananlysis

```python
sns.pairplot(df)
```
